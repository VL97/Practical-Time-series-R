```r
set.seed(2016)
N=1000
phi = .4
Z = rnorm(N,0,1); X=NULL; X[1] = Z[1];
```

Since X[1] = Z[1]; we start our process too from a random value sampled from a normal distribution.

As we know that AR with phi<1 satisfies stationarity condition hence below time series is stationary.

```r
for (t in 2:N) {
X[t] = Z[t] + phi*X[t-1] ;
}
X.ts = ts(X)
par(mfrow=c(2,1))
plot(X.ts,main="AR(1) Time Series, phi=.4")
X.acf = acf(X.ts, main="AR(1) Time Series, phi=.4")
```

If phi=1 (or >1) we violate the stationary condition. We already know that phi=1 gives a random walk which is non-stationary (weak):

```r
phi=1
```

```r
for (t in 2:N) {
X[t] = Z[t] + phi*X[t-1] ;
}
X.ts = ts(X)
par(mfrow=c(2,1))
plot(X.ts,main="AR(1) Time Series (or a Random Walk), phi=1")
X.acf = acf(X.ts, main="AR(1) Time Series (or a Random Walk), phi=1")
```

```r
phi1 = .7; phi2 = .2;
X.ts <- arima.sim(list(ar = c(phi1 , phi2)), n=1000)
par(mfrow=c(2,1))
plot(X.ts,main=paste("AR(2) Time Series, phi1=",phi1, "phi2=", phi2))
X.acf = acf(X.ts, main="Autocorrelation of AR(2) Time Series")
```

```r
phi1 = .5; phi2 = -.4;
X.ts <- arima.sim(list(ar = c(phi1 , phi2)), n=1000)
par(mfrow=c(3,1))
plot(X.ts,main=paste("AR(2) Time Series, phi1=",phi1, "phi2=", phi2))
X.acf = acf(X.ts, main="Autocorrelation of AR(2) Time Series")
X.pacf = acf(X.ts, type='partial', main="PACF of AR(2) Time Series")
```
