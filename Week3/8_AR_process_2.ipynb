For AR(p) of any order p (though we have used AR(1) as examples)

ACF is: 

<h1><center> ùúå(ùëò)=(ùúô1)<sup>ùëò</sup> </center></h1>

```r
set.seed(2016); N=1000; phi = .4;
Z = rnorm(N,0,1)
X=NULL;
X[1] = Z[1];
for (t in 2:N) {
X[t] = Z[t] + phi*X[t-1] ;
}
X.ts = ts(X)
X.acf = acf(X.ts, main='AR(1), phi=0.4')
```

and for AR(p), mean is 0 and variance as:

<h1><center> ùúé<sub>ùëç</sub><sup>2</sup>Œ£ùúô<sub>ùëñ</sub><sup>2</sup> </center></h1> where i ranges from 0 to infinity

```r
var(X.ts)
```

```r
mean(X.ts)
```

Both above values are in agreement

# Some other ùúô values also indicate geometric fall in ACF: 

```r
phi=0.9
for (t in 2:N) {
X[t] = Z[t] + phi*X[t-1] ;
}
X.ts = ts(X)
X.acf = acf(X.ts, main='AR(1), phi=0.9')
```

```r
phi=-0.8
for (t in 2:N) {
X[t] = Z[t] + phi*X[t-1] ;
}
X.ts = ts(X)
X.acf = acf(X.ts, main='AR(1), phi=-0.8')
```

```r
acf(X.ts, type='partial',main='PACF (AR(1), phi=-0.8')
```

clearly PACF can help in true identification of order of AR ie 1 here as seen from PACF.

Differentiate between MA(q) and AR(k):
<li>MA has abrupt ACF cut off after q lags while for AR it follows geometric die out. <li>Further AR's PACF cuts off after k lags.

## We also have geometric fall in ACF for AR(p) p>1

Below is stationary since roots are 2,2 

```r
set.seed(2016); N=1000;
Z = rnorm(N,0,1)
X=NULL;
X[1] = Z[1];
X[2] = Z[2];
for (t in 3:N) {
X[t] = Z[t] + X[t-1]- 0.25*X[t-2];
}
X.ts = ts(X)
X.acf = acf(X.ts, main='AR(2)')
```

Below has roots 2,-1. Note the ACF declines slowly when unit root is present (ie on the unit circle)

```r
set.seed(2016); N=1000;
Z = rnorm(N,0,1)
X=NULL;
X[1] = Z[1];
X[2] = Z[2];
for (t in 3:N) {
X[t] = Z[t] -0.5*X[t-1]+0.5*X[t-2];
}
X.ts = ts(X)
X.acf = acf(X.ts, main='AR(2)')
```
