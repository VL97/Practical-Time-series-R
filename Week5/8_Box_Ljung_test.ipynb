```r
X=rnorm(100000,0,1)
X=ts(X)
plot(X, xlim=c(1,100), col='red', main='White Noise')
```

```r
m=100000
phi=c(.7, .2)
beta=0.5
Simulated.Arima=arima.sim(n=m,list(order = c(2,1,1), ar = phi, ma=beta))
plot(Simulated.Arima,xlim=c(1,10000), col='green', main='An ARMA process')
```

### Box test for White Noise

```r
Box.test(X, lag=log(length(X)))
```

As we have high p value (say 5% level of significance) we fail to reject the NULL hypothesis. Thus the NULL hypothesis that all ACFs for lags upto 11 are all equal to zero holds for WHITE NOISE. This tells us that the series is normally distributed.

### Box test for ARMA

```r
Box.test(Simulated.Arima, lag=log(length(Simulated.Arima)))
```

As we have low p value (say 5% level of significance) we reject the NULL hypothesis. Thus we have at least one non zero ACF till 11 lags. This tells us that the series is not normally distributed.
