# PACF

PACF for AR(p) process cuts off after p lags (theoretically)

```r
phi1 = .6; phi2 = .2;
X.ts <- arima.sim(list(ar = c(phi1 , phi2)), n=500)
par(mfrow=c(3,1))
plot(X.ts,main=paste("AR(2) Time Series, phi1=",phi1, "phi2=", phi2))
X.acf = acf(X.ts, main="Autocorrelation of AR(2) Time Series")
X.pacf = acf(X.ts, type="partial", main="Partial Autocorrelation Function")
```

```r
phi1 = .9; phi2 = -.6; phi3 = .3;
X.ts <- arima.sim(list(ar = c(phi1 , phi2, phi3)), n=500)
par(mfrow=c(3,1))
plot(X.ts,main=paste("AR(3) Time Series, phi1=",phi1, "phi2=", phi2, "phi3=", phi3))
X.acf = acf(X.ts, main="Autocorrelation of AR(3) Time Series")
X.pacf = acf(X.ts, type="partial", main="Partial Autocorrelation Function")
```

# Beveridge wheat dataset

```r
beveridge = read.table("beveridge_wheat.txt", header=TRUE)
#take column no 2 ie Price
beveridge.ts = ts(beveridge[,2], start=1500)
plot( beveridge.ts, ylab="price", main="Beveridge Wheat Price Data")
beveridge.MA = filter(beveridge.ts, rep(1/31, 31), sides = 2)
lines(beveridge.MA, col="red")
```

```r
par(mfrow=c(3,1))
#scale down the series by the MA
Y = beveridge.ts/beveridge.MA
plot( Y, ylab="scaled price", main="Transformed Beveridge Wheat Price Data")
acf(na.omit(Y),main="Autocorrelation Function of Transformed Beveridge Data")
acf(na.omit(Y), type="partial",main="Partial Autocorrelation Function of Transformed Beveridge Data")
```

```r
ar(na.omit(Y), order.max = 5)
```

Hence selected order is 2 for the scaled process
