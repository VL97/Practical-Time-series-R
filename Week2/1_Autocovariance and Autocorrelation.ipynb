# Noise and Sound from ACF

we conclude that ACFs not only depend on a lag on X axis but also on its previous lags.

```r
x = seq(1, 100, by=1)
data = rnorm(100,0,1)
(acf(2+3*x + data, main="ACF: Signal+Noise", col='blue'))
```

ACFs only and only depend on its current lag value on the X axis for noise:

Notice also that none of the ACF is significant: particular characteristic of noise.

```r
data = rnorm(100,0,1)
(acf(data, main="ACF: Noise", col='red'))
```
