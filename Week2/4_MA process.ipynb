```r
# Generate noise
noise=rnorm(1000)
```

```r
# Introduce a variable
ma_2=NULL

# Loop for generating MA(2) process
# must start from i=3 due to order of MA as 2
for(i in 3:1000){
    ma_2[i]=noise[i]+0.7*noise[i-1]+0.2*noise[i-2]
}
```

Notice these coefficients make MA process non invertible

```r
# Shift data to left by 2 units
moving_average_process=ma_2[3:1000]

# Put time series structure on a vanilla data
moving_average_process=ts(moving_average_process)
```

```r
# Partition output graphics as a multi frame of 2 rows and 1 column
par(mfrow=c(2,1))
```

```r
# plot the process and plot its ACF
plot(moving_average_process, main='a MA(2) process', ylab=' ', col='blue')
acf(moving_average_process, main='Correlogram of a MA(2)', col='orange')
```

Note that ACF will be insignificant after q=2 lags. Theoretically, it is 0 post lag 2.
