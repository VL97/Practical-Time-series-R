```r
x=NULL
x[1]=0
for(i in 2:1000){
    x[i]=x[i-1]+rnorm(1)
}
```

```r
random_walk=ts(x)                            #transform to timeseries object
```

```r
plot(random_walk, main='A Random walk', ylab='Yt', xlab='t', col='green', lwd=2)
```

From the ACFs we can see that the random walk is not a stationary series (though ACFs are defined for stationary series):

```r
acf(random_walk, col='blue', main='a Random walk\'s correlogram')
```

We can obtain stationary purely random process from Random walk by first differences:

```r
differ=diff(random_walk)                      #gives difference at lag1
plot(differ)
```

The ACFs specify that differ gives stationary series:

```r
acf(differ, col='red',main='Correlogram of difference')
```
